# Regression

```python
ax=recent_grads.plot(x="Sample_size",y="Median",kind="scatter",label="Sample_size and Median")
```

```python
from scipy.stats import linregress
x=recent_grads["Sample_size"]
y=recent_grads["Median"]
(a,b,_,_,_)=linregress(x,y)
```

```python
y_new=a*x+b
```

```python
fig,ax=plt.subplots()
ax.scatter(x,y)
ax.plot(x,y_new,"b--", color="red")
```
